```python
#Seaborn is the statiscal plotting library
#It has beautiful styles
#It is also designed to work very well with pandas dataframe objects

#categorical plots
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)#shows categorical distrbution
```

```python
sns.countplot(x='sex',data=tips)#works as a count in pandas
```

```python
#This type of plot is most of the times is used to  show the categorical data also known as whisker plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind="violin")
```
