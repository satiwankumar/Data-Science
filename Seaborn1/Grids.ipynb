```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
import matplotlib.pyplot as plt
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)#pair the numerical columns 
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
```
