```python
import pandas as pd 
import numpy as np
import os
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
#categorcal vaiable where the value doesnot have any categorical ordering
#green,red,blue
#ordinal where categories have some sort of numeric ordering btw them
#graduate,masters,phd,high ,medium ,low
```

```python
dummies= pd.get_dummies(df.town)
```

```python
merged  = pd.concat([df,dummies],axis ='columns')
merged
```

```python
#Rule of the Thumb
#we have to drop one varaible becuase whe u have mutiple variable driving from one variable it will mess machine leaning model
#hence drop one colinear vairbale from dataset

final = merged.drop(['town','west windsor'],axis = 'columns')
final
```

```python
model = linear_model.LinearRegression()
```

```python
x = final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
