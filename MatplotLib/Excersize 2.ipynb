```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Matplot is the most popular plotting library for python
#IT gives you control over eery aspect of figure
x=np.linspace(0,5,11)
x
y = x**2
y
```

```python
#function 
plt.plot(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("line chart")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented API method
fig = plt.figure()

axes = fig.add_axes([0.8,0.8,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Chart')

axes1 = fig.add_axes([0.4,0.2,0.8,0.8])
axes2 = fig.add_axes([0.5,0.5,0.3,0.4])
axes1.plot(x,y)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Chart')
axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Chart')
```

```python
#subplots  
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First plot ')
axes[1].plot(y,x)
axes[1].set_title('second title')
plt.tight_layout()
```

```python
#figure size and DPI
fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y ,label="x,y")
ax.plot(y,x,label="y,x")
ax.legend(loc=10)# location will define where to show legend
```
