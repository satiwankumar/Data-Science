```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
expense_value = [1400,1600,300,400,150]
exp_labels = ["Home","Rent","Food","Phone","Other Utilities"]
```

```python
plt.pie(expense_value,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.savefig('tessstttyyy.png', dpi=100)
plt.show()
```
