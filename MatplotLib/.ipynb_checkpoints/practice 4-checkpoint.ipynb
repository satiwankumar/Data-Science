```python
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
blood_sugar = [113,85,90,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,155,84,89,120,112,100]
```

```python
plt.hist([blood_sugar,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['orange','green'],label=['man','women'],orientation="horizontal")
plt.xlabel=('sugar range')
plt.ylabel=("total number of patients")
plt.title("Blood Sugar Analysis")
plt.legend()
```
