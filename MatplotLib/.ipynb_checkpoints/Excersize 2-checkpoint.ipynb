```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Matplot is the most popular plotting library for python
#IT gives you control over eery aspect of figure
x=np.linspace(0,5,11)
x
y = x**2
y
```

```python
#function 
plt.plot(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("line chart")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented API method
fig = plt.figure()

axes = fig.add_axes([0.8,0.8,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Chart')
```
